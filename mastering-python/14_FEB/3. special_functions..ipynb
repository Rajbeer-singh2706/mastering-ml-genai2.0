## **filter()**

```python
# Filter(function, iterable)
# function that take one item and return a boolean

list_a= [1,2,3,4,5]

def is_even(num):
    return num%2 == 0


result = list(filter(is_even, list_a))
result
```

## **List & Dict COmpreshension**

```python
numbers = list(range(1,21))
numbers
```

```python
def even_numbers(l):
    final_even_list = []
    for i in l:
        if i%2 ==0:
            final_even_list.append(i)
    return final_even_list
```

```python
even_n = even_numbers(numbers)
even_n
```

## **list compreshension**

```python
even_n = [item for item in numbers if item%2 == 0 ]
even_n
```

```python
even_n = [item  if item%2 == 0 for item in numbers]
even_n
```

```python
even_n = [item  if item%2 == 0 else -1 for item in numbers]
even_n
```

```python
## Problem Statement 
numbers = range(1,101)
is_even_list_v1 = [True if item%2 else False for item in numbers ]
is_even_list_v1 = [True if item%2 else False for item in range(1,101) ]
is_even_list_v1
```

```python
numbers = range(1,101)
is_even_list_v2 = [ item%2 ==0 for item in numbers ]
is_even_list_v2
```

## **dictionary comprehension**

```python
even_dict = {item:item for item in numbers if item%2 ==0 }
even_dict
```

```python
even_dict = { num:True if num%2 == 0 else False for num in range(1,100)}
even_dict
```
