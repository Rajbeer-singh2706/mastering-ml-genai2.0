```python
class Wallet:
    def __init__(self, cash):
        self.cash = cash
    
    def __repr__(self) -> str:
        return f"class Wallet, cash: {self.cash}"
    
    ## Special Method
    def __add__(self, other) -> int:
        if(isinstance(other, Wallet)):
            return Wallet(self.cash + other.cash)
        else:
            #raise "please provide Wallet Object"
            return None
    
    ## Special method
    def __sub__(self, other):
        return Wallet(self.cash - other.cash) 

    def __eq__(self, other ):
        return self.cash == other.cash 
```

```python
w1 = Wallet(50)
w2 = Wallet(30)

result = w1 + w2 
print(result)
print(result.cash)
```

```python
### Addition 
w1 = Wallet(50)
w2 = Wallet(30)

result = w1 + w2 
print(result)
```

```python
### Subtraction
w1 = Wallet(50)
w2 = Wallet(60)

result = w1 - w2 
print(result)
```

```python
### Equality 
w1 = Wallet(50)
w2 = Wallet(50)
 
print(w1 == w2 )
```

## **Next Specila method: def `__getitem__`(self, index):**

```python
class ATM:
    def __init__(self):
        self.transaction_history = []
    
    def __repr__(self):
        return f"Class ATM, Cash "
    
    def do_transaction(self,value):
        self.transaction_history.append(value)
        print(f"Transaction done : {value}")
    
    def __len__(self):
        return len(self.transaction_history) 
    
    def __getitem__(self, index):
        return self.transaction_history[index]
```

```python
atm_2 = ATM()
```

```python
atm_2.do_transaction(100)
```

```python
atm_2.do_transaction(100)
```

```python
atm_2.do_transaction(-100)
```
