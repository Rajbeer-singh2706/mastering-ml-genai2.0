# Special Methods (OOPS)

```python
class Car:
    def __init__(self,manuf, car_name):
        self.manuf = manuf
        self.car_name = car_name

    def __str__(self):
        return f"Manufacturer Name : {self.manuf} and we produce car."
    
    def __repr__(self): # official representation of an object
        return f"Class Car, Manufacturer Name : {self.manuf} and we produce car. This object contains the car model : {self.car_name}"
```

# Property decorator

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance

    # getter
    # def get_balance(self):
    #     return self.__balance

    # setter
    # def set_balance(self, value):
    #     self.__balance = value

    # get, set, del

    # getter
    @property
    def balance(self):
        print("Checking balance")
        return self.__balance
    
    # setter
    @balance.setter
    def balance(self, value):
        if value < 0:
            print("Security Alert. balance cannot be negative")
        elif value > 1000000:
            print("Security Alert. large transaction, requires approval")
        else:
            print("Changing balance")
            self.__balance = value

    @balance.deleter
    def balance(self):
        print("ERROR : You cannot delete a bank balance.")
    
```

### filter

```python
# filter(function, iterable)
# function that takes one item and return a boolean
list_a = [1,2,3,4,5]

def is_even(num):
    # 2 --> 2%2==0 -> True
    # 3 -> 3%2==0 -> False
    return num % 2 == 0
```

```python
list(filter(is_even, list_a))
```

```python
numbers = list(range(1,21))
print(numbers)
```

```python
def even_numbers(l):
    final_even_list = []
    for i in l:
        if i%2==0:
            final_even_list.append(i)
    return final_even_list
```

```python
even_n = even_numbers(numbers)
```

```python
print(even_n)
```
