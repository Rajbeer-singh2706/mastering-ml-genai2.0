```python
class A:
    def __init__(self):
        print("A()")

    def speak(self):
        print("Parent Speak()")

class B(A):
    def __init__(self):
        super().__init__()
        print("B()")

    def speak(self):
        super().speak()
        print("Child Speak()")
```

```python
object1 = B()
object1.speak()
```

```python
class MotherBoard:
    def __init__(self, ram, cpu_speed, cpu_type):
        self.ram = ram 
        self.cpu_speed = cpu_speed
        self.cpu_type = cpu_type
    
    def start(self):
        print("Power flow started")
    
    def stop(self):
        print("Power flow stopped")

class Computer(MotherBoard):
    def __init__(self, ram, cpu_speed, cpu_type):
        super().__init__(ram, cpu_speed, cpu_type)
        #MotherBoard(ram, cpu_speed, cpu_type)
        self.os = self.__check_os_and_set_os_type()

    def __check_os_and_set_os_type(self): # Private Method
        if self.cpu_type.lower() == "mac-chip":
            return "mac"
        else:
            return "other"
    
    def check_my_os(self): # public Method 
        print("user Has : ", self.os, "OS")
```

```python
computer_object1 = Computer("12GB","4.2ghz","mac-chip")
computer_object1.check_my_os()
computer_object1.start()
computer_object1.stop()
```

```python
from abc import ABC, abstractmethod

## ABC ==> abstract class 
```

```python
## I want to defina a contract
class Payment(ABC): # requirement no 1 

    @abstractmethod # decorator
    def pay(self, amount , currency):
        pass 
    
    @abstractmethod
    def refund(self, amount):
        pass 

# Contact called payment has 2 methods in the contract which all baks needs to implement
# First is pay 
# second is refund 
```

```python
class TestBank(Payment):
    def new_user(self):
        print("New User") 
```

```python
object1 = TestBank()
object1.pay()
```
