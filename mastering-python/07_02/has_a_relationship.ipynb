## **1. Simple Composition(Has a via Instance Variable)**

```python
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        return "Engine Started"

class Car:
    def __init__(self, brand, horsepower):
        self.brand = brand 
        self.engine = Engine(horsepower=horsepower) # CAR HAS AN Engine
    
    def drive(self):
        return self.engine.start() + " - Car is driving"
    
# Usage 
car = Car("Toyota", 150)
car.drive()
```

## **2. Composition with INitialization**

```python
class Address:
    def __init__(self, street, city):
        self.street = street
        self.city = city

class Person:
    def __init__(self, name, street, city):
        self.name = name
        self.address = Address(street, city)  # Person HAS AN Address
    
    def get_location(self):
        return f"{self.name} lives in {self.address.city}"

# Usage
person = Person("John", "123 Main St", "New York")
print(person.get_location())
```

## **3. Composition with Dependency Injection**

```python
class Wheel:
    def __init__(self, size):
        self.size = size

class Bike:
    def __init__(self, brand, wheel):
        self.brand = brand
        self.wheel = wheel  # Pass object as parameter
    
    def get_specs(self):
        return f"{self.brand} bike with {self.wheel.size} inch wheels"

# Usage
wheel = Wheel(26)
bike = Bike("Giant", wheel)
print(bike.get_specs())
```

## Key differences
 * Composition: The contained object's lifecycle is tied to the parent (often created inside)
 * Aggregation: The contained object can exist independently (often passed in)
