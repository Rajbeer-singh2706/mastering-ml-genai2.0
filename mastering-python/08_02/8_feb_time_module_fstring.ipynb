```python
t1_perf = time.perf_counter()
add(20,20)
t2_perf = time.perf_counter()
print(" Time taken : ", (t2_perf- t1_perf), "len : ", len(str((t2_perf-t1_perf))))
```

## **time.sleep()**

```python
add(10,20)
time.sleep(5) # wait for 5 second
add(10,20)
```

```python
t1 = time.time()
time.sleep(5)
t2 = time.time()
print(" Time taken : ", (t2- t1), "len : ", len(str((t2-t1))))
```

```python
t1_perf = time.perf_counter()
time.sleep(5)
t2_perf = time.perf_counter()
print(" Time taken : ", (t2_perf- t1_perf), "len : ", len(str((t2_perf-t1_perf))))
```

## **divmod**

```python
# HH: MM: SS 
# divmod --> div + mod 

hrs , remainder = divmod(t3-t1, 3600)
minutes, seconds = divmod(remainder, 60 )

print(" HRS : ", hrs, "remainder : ", remainder)
print(" HRS : ", hrs, "minutes : ", minutes, "seconds", seconds)
```
