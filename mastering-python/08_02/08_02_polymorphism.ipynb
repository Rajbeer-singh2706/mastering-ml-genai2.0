## **With Polymorphism**

```python
## lets create a contract 
## ABSTRACTION 
from abc import ABC, abstractmethod
class ElectronicContract(ABC):
    @abstractmethod
    def start(self): pass 
```

```python
### With PolyMorphisem 
class TV: 
    def start(self):
        print("TV")

class MicroWave: 
    def start(self):
        print("MicroWave")

class FAN: 
    def start(self):
        print("FAN")
```

```python
electronic_appliances = [TV() , MicroWave() , FAN() ]
electronic_appliances
```

```python
for e in electronic_appliances:
    e.start()
```

```python
### With PolyMorphism + ABSTRACTION 
class TV(ElectronicContract): 
    def start(self):
        print("TV")

class MicroWave(ElectronicContract): 
    def start(self):
        print("MicroWave")

class FAN(ElectronicContract): 
    def start(self):
        print("FAN")

class AirPurifier(ElectronicContract):
    # def start_airfried(self):
    #     print("start airpurifier")

    def start(self):
        print("start airpurifier")
```

```python
electronic_appliances = [TV() , MicroWave() , FAN(), AirPurifier() ]
electronic_appliances

# TypeError: Can't instantiate abstract class AirPurifier without an implementation for abstract method 'start'
```

```python
for e in electronic_appliances:
    e.start() 
```
