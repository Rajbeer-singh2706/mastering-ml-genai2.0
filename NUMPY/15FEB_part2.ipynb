```python
import time 
```

```python
start_e = time.perf_counter()
x = np.empty((5,5), dtype = np.float16)
#x = np.empty((5,5), dtype = np.float128) # error 
print(x)
end_e = time.perf_counter()
print(end_e-start_e)
```

```python
start_we = time.perf_counter()
x = np.ones((5,5))
print(x)
end_we = time.perf_counter()
print(end_we-start_we)
```

* return the size in bytes of one element 
* 1 byte --> 8 bit 
* int8 --> 8 bit 
* int64 --> 64 bit (8 * 8) # (int8 * 8)
