## рҹҺЁ Decorators

> **Definition:** A Decorator is a **function** that takes **another function** as an **argument** and **returns a new function**. It is used to **enhance the functionality** of the base function.

### Diagram:
```
          в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ
          в”Ӯ function в”Ӯ
          в””в”Җв”Җв”Җв”Җв”¬в”Җв”Җв”Җв”Җв”Җв”ҳ
               в”Ӯ (passed as argument)
               в–ј
        function(  )  в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҖвҶ’  new function
```

### How it works вҖ” Step by step:

#### Step 1 вҖ” Base function:
```python
def add():
    a = 1        # в‘   вҶ’ return should be present here
    return a     # в‘Ў
```

#### Step 2 вҖ” Decorator function:
```python
def run(func):          # в‘ў
    def wrap():         # в‘Ў
        print()         # в‘ 
        func()
        print()
    return wrap   вҶ’ в‘ 
    # return run  вҶ’ в‘Ў
```

## рҹ“„ Decorators вҖ” Full Code Example (Page 2)

```python
def add():
    print("Addition Function")       # definition вҶ’ memory address: 101

def my_func(func):
    def wrap():
        print("Start")
        func()
        print("End")
    return wrap

add = my_func(add)

add()
```

### Execution trace when `add()` is called:

```
add  вҶ’ 101
func вҶ’ 101

my_func(func)
  def wrap():
    print()      вҶҗ "Start"
    func()       вҶ’ 103
    print()      вҶҗ "End"
  return wrap
```

### Key insight:
```
add = my_func(add)
         вҶ‘
       вқҢ NOT 101 anymore
       вң… address of a function called as wrap вҶ’ 103
```

So now when you call `add()` вҶ’ it actually runs `wrap()` at address 103.

### Static вҖ” Definition & Properties:

> **Static = belong to the CLASS, not an object**

```
вңҰ Shared across all objects
вңҰ Created once
вңҰ Not tied to any instance
```


### Class diagram:

```python

```

