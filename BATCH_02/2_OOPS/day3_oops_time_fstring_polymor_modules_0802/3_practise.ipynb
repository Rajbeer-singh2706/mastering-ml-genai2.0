```python
class TV:
    def start_tv(self):
        print("TV")
    
class Microwave:
    def start_mv(self):
        print("MV")

class Fan:
    def start(self):
        print("Fan")
```

```python
tv = TV()
tv.start_tv()

mv = Microwave()
mv.start_mv()

f = Fan()
f.start()
```

```python
from abc import ABC, abstractmethod
```

```python
class ElectronicContract(ABC):
    @abstractmethod
    def start(self):
        pass 
```

```python
# Polymorphism concept

class TV(ElectronicContract):
    def start(self):
        print("TV")

class Microwave(ElectronicContract):
    def start(self):
        print("MV")  
    
class Fan(ElectronicContract):
    def start(self):
        print("Fan")  

class AirFrier(ElectronicContract):
    def start_airfried(self):
        print("start_airfried")  

    def start(self):
        print("AirFrier")  
```

```python
electronic_appliances = [TV(), Microwave(), Fan(), AirFrier()]
```

```python
for e in electronic_appliances:
    e.start()
```
