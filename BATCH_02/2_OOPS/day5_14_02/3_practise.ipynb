```python
class Wallet:
    def __init__(self, cash):
        self.cash = cash 
    
    def __repr__(self):
        return f"Class Wallet, Cash: {self.cash}"
    
    def __add__(self, other):
        return Wallet(self.cash + other.cash)
```

```python
w1 = Wallet(50)
w2 = Wallet(30)

combined_sum = w1 + w2 
print(combined_sum.cash)
```
