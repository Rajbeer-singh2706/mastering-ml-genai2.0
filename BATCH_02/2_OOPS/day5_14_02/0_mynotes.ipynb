## ➕ `__add__` Dunder Method

### Concept diagram:
```
A → wallet(50) → w1
B → wallet(30) → w2

  (w1)  +  (w2)  →  80 (wallet)
   ↑           ↑
  obj          obj
        ↓
      __add__
         ↑
   w1.__add__( w2 )  →  80 (wallet)
              ↑
        self.cash + other.cash
```


## 📄 `__add__` — Full Execution (Page 2)

```
obj = w1 + w2
obj.cash → 80

w1.__add__(other):        ← w2
    wallet(self.cash + other.cash)
          50        30
    return wallet(80)
              ↓
        creating a new object
        with self.cash = 80
```

### Commutativity:
```
w1 + w2  →  ①   calls  w1.__add__()
w2 + w1  →  ②   calls  w2.__add__()
```

### Float vs Int behavior:
```
w1.cash = 30
w2.cash = 40.3

70.3  → float ✅
70    → int

1 + 1      → 1 (int)
1 + 1.3    → 2.3 (float)

float( __  +  __ )
int  ( __  +  __ )
```


## 🏦 Property Decorator — `@property` (Page 3)

### Problem without property:
```python
class Account:
    def __init__(self, balance):
        self.balance = balance

    def set_balance(value):
        self.balance = value
```

```
a = Account(1000)
        → self.balance = 1000

a.set_balance(2000)
        → self.balance = 2000

a.balance = 3000   ← # security issue ⚠️
```


## 🔍 `__name__` Special Variable

### Module diagram:
```
my_module
  └── test_1.py
        └── print(__name__)

main.py
  └── from my_module import test_1
        └── print(__name__)
```

### Execution flow:
```
① python main.py → test_1 → print(__name__)  → def / class code → execute
②                 print(__name__)
```

> **`__name__` is a special variable — it holds its execution file name in it.**


## 📄 `__name__` Rules (Page 4)

### Rule ①:
```
If __name__ is present in the SAME file you are executing:
    __name__  =>  "__main__"
```

### Rule ②:
```
If __name__ is NOT present in the current file you are executing:
    __name__  →  that file's name
```

## ⚡ Special Functions

### Lambda Function — One-line function:

```python
# Normal function:
def add(a, b):
    return a + b

add(1, 2) → 3

# Lambda equivalent:
add = lambda a, b: a + b

add(1, 2) → 3
```

## 📋 List Comprehension

### Evolution of list comprehension:

```python
# loop/iteration → value one by one
→ [i for i in numbers]          # iterate → ❌ get → [ ]

[ for i in numbers ]            # iterate → get → [ ]

[ i for i in numbers ]          # ✅ iterate → collect into list
        ↑
     iterate
```

### With condition (filter):
```python
[num for num in numbers if num % 2 == 0]
      ①                    ② check → if true
          ↓
       [ num ]
```

```
① → loop through numbers
② → check condition → if True → add to list
```
