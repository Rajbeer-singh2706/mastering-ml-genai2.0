```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name # public 
        self._account_type = "Savings" # Protected
        self.__balance = balance # Private
    
    def get_balance(self):
        # getter 
        # to get value of the balance variable 
        print("Your current balance :", self.__balance)
    
    def add_and_set_new_balance(self,amount):
        # setter 
        # to deposit money in our account
        self.__balance +=amount 
        print("Your current balance: ", self.__balance)
```

```python
class MotherBoard:
    def __init__(self, ram, cpu_speed, cpu_type):
        self.ram = ram 
        self.cpu_speed = cpu_speed 
        self.cpu_type = cpu_type 
    
    def start(self):
        print("Power flow started")
    
    def stop(self):
        print("Power flow stopped")
```

```python
class Computer(MotherBoard):
    def __init__(self, ram ,cpu_speed, cpu_type):
        super().__init__(ram, cpu_speed,cpu_type)
        self.os = self.__check_os_and_set_os_type()
    

    def __check_os_and_set_os_type(self): # Private method 
        if self.cpu_type.lower() == "mac-chip":
            return "mac"
        else:
            return "other"
    
    def check_my_os(self):
        print("User has: ", self.os , "OS")
```

```python
cmp1 = Computer("12gb", "4.2ghz", "mac-chip")
```

```python
cmp1.check_my_os()
```

```python
cmp1.start()
```

```python
cmp1.stop()
```

```python
class A:
    def __init__(self):
        print("Class A")
    
    def a(self):
        print("a")


class B(A):
    def __init__(self):
        super().__init__()
        print("Class B")
    
    def b(self):
        print("b")


class C(A):
    def __init__(self):
        print("Class C")
    
    def c(self):
        print("c")

class D(B,C): # not available in java # ambiguity 
    def __init__(self):
        super().__init__()
        print("Class D")
    
    def d(self):
        print("d")
```
