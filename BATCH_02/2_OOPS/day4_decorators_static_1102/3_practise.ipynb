```python
class MathUtils_New:
    
    name = "UDSB2"

    @staticmethod
    def add(a,b):
        return a + b 

    @staticmethod
    def subtract_a_from_b(a,b):
        return a - b 

    @staticmethod
    def subtract_b_from_a(a,b):
        return b - a

    def class_method(self):
        print(self.add(1,2))
        print(MathUtils_New.add(1,2))
```
